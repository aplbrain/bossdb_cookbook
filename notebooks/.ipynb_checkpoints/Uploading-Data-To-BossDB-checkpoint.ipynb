# CON2021 Notebook 2: bossDB Advanced User Tutorial

Now that we've gone through some basics on downloading data and visualizing it, we can get to some more advanced topics on creating and uploading data. 

There are two main ways to upload data onto bossDB:

1. Ingest-client: A python-based client that can be used to transfer arbitrary file formats (PNGs, TIFFs, Zarr) into the bossDB. Its highly scalable (uses SQS queues) and works well with parallel infrastructure. There is some set-up involved however and requires a little more hands-on experience. (https://github.com/jhuapl-boss/ingest-client)

2. Intern's create_cutout(): Function within `intern` that uses bossDB's RESTful API to send data through HTTP. Works only with numpy arrays. Much less set-up but not efficient for large uploads. 


Typically we use intern for anything smaller than 25 GB of data. If we are uploading anything larger than that, we set up and ingest job using the ingest client to handle the upload. In this notebook, I'll mainly be going through uploading data through intern.

## Setting up resources

Just like how downloading data needed a collection, experiment, and channel to download data, in this case we need to create a collection, experiment, and channel to upload data. 

You can do this through `intern` directly or through the management console. If its a single or only a couple of channels, I usually go through the managament console website since its a little easier to parse, but I'll show both methods here!

### Step 1: Set up collection

This is basically the highest level "directory" of your data. Its just a name-space that you reserve like your lab name, dataset and/or project name. In this case, we can create a collection named **CIRCUIT_CON2021** for all of our generated data channels. 

To do this, navigate to the management console website and log in: https://api.bossdb.io/

Once there, you can click "Manage Resources" and then "Add Collection". You'll see a modal like this pop up:



You can then input a **unique** collection name and optional description. Emphasis on the unique, no two collections can share the same name! Since I've created the CIRCUIT_CON2021 collection you do not have to create a collection for these examples. 

Once you've created the collection, click the "Details" button in the Resources table. (You might have to filter by the name or scroll through some pages to find the collection you created). In the details page you will say a table with the experiments available, which should be empty since the collection was just created. Scroll down to permissions. Here you should see admin access for the account that created the collection (presumably yours) as well as other admins. You can add read (or write) access to other groups by clicking "Add Permissions" and then inputting the name of the group to add permissions for as well as the type of permission. 




In this case I've created a group for your cohort: **CON2021** with admin rights to this collection. Before we move on to creating an experiment, we first need to create a coordinate frame. 

### Step 2: Set up coordinate frame

Every experiment in bossDB is attached to a coordinate frame. A coordinate frame is not a directory-like object but instead meta-data that specifies the X,y,z voxel extents and size. Coordinate frames are always public and can be used across multiple experiments. Since coordinate frames exist in the experiment level, all channels within the same experiment share the same coordinate frame. This is especially important for aligning EM with segmentation or other data derivatives. 

To create a coordinate frame, navigate back to the "Manage Resources" tab. If you scroll down you will see a list of coordinate frames. You can add one by clicking the "Add Coordinate Frame" button which will open a modal with a form. Most of the fields are pretty self-explanatory but there are a few quirks not mentioned directly on the form:

1. Heavily recommend that x,y,z starts are all set to 0. Having a non-zero start can complicate downsample and cuboid-alignment.
2. It is also recommended that x,y,z stops are cuboid aligned. The bossDB standard cuboid size is 512,512,16 in XYZ. This means that X and Y stops are multiples of 512 while z stop is a multiple of 16. 
3. Voxel resolution and units do matter for neuroglancer. 
4. Coordinate frames are cheap. It doesn't create any "data", it just marks what areas in the XYZ spatial domain data does exist. So if you don't know exactly how much data you are going to process, its probably better to make the coordinate frame larger than you expect. Once the XYZ values are inputted and submitted, they are difficult to change. 



Once you've set up your coordinate frame, you can go back to collection details from the step before and add an experiment. 

### Step 3: Set up experiment

Similar to how we set up the collection, we click "Add Experiment" button on the collection details page to add an experiment. Here there are more input fields to enter. 



The experiment name is unique to the specfic collection you're in. The coordinate frame input needs to be the name of the coordinate frame you created or exists already. Num hierarchy levels lets bossDB know how many levels of downsample this dataset should have. The hierarchy method is downsample method: anistropic or isotropic. (I'll talk about this more some other time.) 

For the most part we are working with volmetric data with no additional time samples so we can leave num time samples to 1 and everything else relating to time samples blank. 

Similar to collection, you can click the "Details" button on the experiment and scroll down to permissions. Here you should add read/write acess to your group, **CON2021**, so they can also read/write to your newly created experiment. 

### Step 4: Setting up Channel

Last step! Once we've created an experiment and collection, we can finally add channels which actually CONTAIN the volumetric data. To create a channel, go to the experiment details page and click the "Add Channel" button. A modal with several fields should appear prompting you for channel name, type, and other information. Note about channel types:

"image" is reserved for 8-bit or 16-bit image data from EM, x-ray, and other modalities. "annotation" type is always 64-bit. 



Base resolution should be 0 in most cases except when you are uploading an already downsampled version of the data. For annotation channels, its help to also specify the source channel if it exists within the same collection. For example, if I have annotation channel of synapse locations based off of image channel "em_chan3" in the same experiment, I would write "em_chan3" as my source chan. 

Once you've created the channel, add read/write permissions by click "Details" next to the channel and "Add permissions". At this point you are ready to upload data to bossDB!

## Uploading data through intern.

We can upload data by pointing the intern array object to our newly created channel. Public accounts do not have permissions to upload data so first we will have to configure intern to use your account token. To do this, navigate to the API Token page on the management console located here: https://api.bossdb.io/v1/mgmt/token

If the token field is blank, you can click "generate token" to create a new one. Your token is specific to your account and essentially is your password. Make sure not to share it publically on github repos! If you do accidently share your token, you can revoke it on the same page where you generated it and generate a new one. 

Once you have your API we can use it directly on intern to authorize your client to upload data. There are multiple ways to do this as described by this wiki: https://github.com/jhuapl-boss/intern/wiki/Boss%20Setup%20Tutorial#configuration-dictionary

For example, lets say my token is `1e7437c3739hfu93n92850275ne222474ce7f6`. On Mac or linux*, I can create a file at `~/.intern/intern.cfg` and input the following block:
```cfg
[Default]
protocol = https
host = api.bossdb.io
token = 1e7437c3739hfu93n92850275ne222474ce7f6
```

\* Need to test for windows users


Once the configuration file is added intern will automatically use the token provided. Another way to configure intern to use your token is to pass it directly into the array constructor:

```python
import intern

array = intern.array("bossdb://mycol/myexp/mychan", 
            boss_config={
                "protocol": "https",
                "host": "api.bossdb.io",
                "token": "1e7437c3739hfu93n92850275ne222474ce7f6"
            })
```

Once you are authorized, uploading data is pretty simple. Here is an example with some random values.

````python
import numpy as np
from intern import array

# Create dataset array object. I already have my token saved in a config file so I don't have to specify it here.
# I specify the resolution I want to upload (res 0) but not necessary for uploads to base resolution, which 
# usually defaults to 0. 

# URI is bossdb://collection/experiment/channel
dataset = array("bossdb://CIRCUIT_CON2021/xenesd1/image", resolution=0)
````

````python
# Lets get an example image from skimage.

from skimage import data
import matplotlib.pyplot as plt


# camera
camera = data.camera()

# Make camera a uint8 array. 
camera = camera.astype('uint8')
plt.imshow(camera, cmap='gray')

# Intern expect an ZYX array but we only have an XY array. 
# We can get it in the correct format by adding an axis at the end and transposing.

camera = camera[np.newaxis, ...]

## Final thing. Arrays need to be C-contiguous, meaning that they are stored in memory in sequential C order.

camera = np.ascontiguousarray(camera)
````

````python
camera.shape
````

````python
# We can upload by setting the array at our desired extents to the data
dataset[0:1,0:512,0:512] = camera
print("Done Uploading Data!")
````

Lets verify the data upload correctly. 

````python
fig, ax = plt.subplots(1,2, figsize=(20,8))

ax[0].imshow(data.camera(), cmap='gray')
ax[0].set_title("Original data")
ax[1].imshow(dataset[0, 0:512, 0:512], cmap='gray')
ax[1].set_title("Data Retrieved from bossDB")
````
